```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def dice():
    return random.randint(1,10)
```

# Showing my die is fair

All the numbers have similar (equal) chances to occur, thus, my die is fair 

```python
test = [dice() for a in range(int(1e5))]
_=plt.hist(test,bins=10)
```

# define what option1 and option2 are

```python
def option_1():
    return dice() + 1 - 2
def option_2():
    return max([dice() , dice()]) - 2
```

# Distribution of option1 and option2. Simulate the game 10,000 times in a roll.
1. option 1 is uniformly distributed
2. because the high roll in two rolls is picked, option 2 is distributed bais to higher values. 

```python
n = 1e5
test_1 = [option_1() for a in range(int(n))]
test_2 = [option_2() for a in range(int(n))]
_=plt.hist(test_1,color='b',alpha=0.5,label='option 1',bins=10)
_=plt.hist(test_2,color='r',alpha=0.5,label='option 2',bins=10)
_=plt.legend()
```

# these are the unique possible outcomes of option 1 and option 2

```python
print('Option 1:',np.unique(test_1),'\nOption 2:',np.unique(test_2))
```

# Let's just play the game and see what the outcomes are
1. Seems like option 2 has higher chance to be higher. 

```python
doTime = 500
fig, ax = plt.subplots(figsize=(15,8))
play=[]
for ii in range(doTime):
    if option_1 == option_2:
        pass
    else:
        play.append(int(option_1() > option_2()))
    
ax.plot(np.arange(len(play)),play)
_=ax.set(ylim=(-1,2),yticks=(0,1),yticklabels=('option 2 has higher score','option 1 has higher score'),
      title='play the game for %d times'%doTime)
```
