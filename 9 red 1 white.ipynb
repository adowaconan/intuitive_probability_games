```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

Given you have 9 reds and 1 white marbles in a bag, you and your 9 friends take out 1 marble out of the bag. Whoever has got the white one wins. Which position has the highest chance to win?

```python
bag = np.concatenate([['red']*29,['white']*1])
bag

results = np.array([np.random.choice(bag,size=bag.shape[0],replace=False) for ii in range(int(1e5))])

fig,ax = plt.subplots(figsize=(8,5))
ax.hist(np.where(results == 'white')[1],bins=range(len(bag)+1),rwidth=.5,normed=True)
#sns.kdeplot(np.where(results == 'white')[1],shade=True,ax=ax,kernel='gau',bw=0.1,cut=0,)
ax.set(xlabel='Position wins',ylabel='Counts',title='time of wins if on different position')
```

What if there are 8 reds and 2 whites?

```python
bag = np.concatenate([['red']*28,['white']*2])
bag

results = np.array([np.random.choice(bag,size=bag.shape[0],replace=False) for ii in range(int(1e5))])
positions = [next((ii for ii,x in enumerate(row) if (x == 'white')), None) for row in results]
fig,ax = plt.subplots(figsize=(8,5))
ax.hist(positions,bins=range(len(bag)+1),rwidth=.5,normed=True)
#sns.kdeplot(positions,shade=True,ax=ax,bw=1,cut=0)
ax.set(xlabel='Position wins',ylabel='Counts',title='time of wins if on different position')
```

```python
def bags(reds = 10,whites = 10, sim_time = int(1e5)):
    bag = np.concatenate([['red']*reds,['white']*whites])
    results = np.array([np.random.choice(bag,size=bag.shape[0],replace=False) for ii in range(sim_time)])
    positions = [next((ii for ii,x in enumerate(row) if (x == 'white')), None) for row in results]
    fig,ax = plt.subplots(figsize=(8,5))
    ax.hist(positions,bins=range(len(bag)+1),rwidth=.5,normed=True)
    #sns.kdeplot(positions,shade=True,ax=ax,bw=1,cut=0)
    ax.set(xlabel='Position wins',ylabel='Probability to win',title='{} whites and {} reds'.format(whites,reds))
    return results,positions
```

```python
results,positions = bags(25,5)
```
