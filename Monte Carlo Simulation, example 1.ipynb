Consider an imaginary game in which our player "jack",rolls an imaginary dice with 100 faces. If Jack lands on 1-51, the house wins, but if Jack lands on 52 - 100, Jack wins. 

# Analytical approach

$$P(Jack wins) = 49/100$$
$$P(casino wins) = 51/100$$
$$\mathbf{E}(profit of Jack)=(1)*(P(\text{Jack wins})+ (-1)*(P(\text{cosino wins}))$$
$$=1*(49/100)-1*(51/100)$$
$$=-0.02$$
$$=-2\%$$

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
"""
create function for simulating die rolling
the die can take values from 1 to 100.
if the number is between 1 to 51, the house wins
else, Jack wins
"""
def rolldice():
    dice = np.random.choice(np.arange(1,101,1),size=1,replace=False)[0]
    
    if dice <= 51:
        return False
    elif (dice > 51) and (dice <= 100):
        return True
```

1. Total funds: the money Jack has is \$10,000
2. Wager Amount: the amount Jack bets in each game \$100
3. Total plays: the number of times Jack plays the game (the value is changed for creating different scenarios)

```python
"""
Define a function for the play which taks 3 arguments:
    1. total_funds
    2. wager_amount
    3. total_plays
"""
def play(total_funds=10000, wager_amount=100,total_plays=100):
    """
    create empty lists for:
    1. play_number
    2. funds available
    3. final fund
    """
    play_num = []
    funds = []
    final_funds = []
    
    
    play_num.append(0)
    funds.append(total_funds)
    # if number of plays is less than the max number of plays we have set
    # I will avoid using while-loops in python
    for play in range(int(1e7)):# just make it a lot
        if rolldice():# if Jack wins
            # add the money to our funds
            total_funds = total_funds + wager_amount
            # append the play number
            play_num.append(play+1)
            # append the new fund amount
            funds.append(total_funds)
        else: # if house wins
            # add the money to house's funds
            total_funds = total_funds - wager_amount
            # append the play number
            play_num.append(play+1)
            # append the new fund amount
            funds.append(total_funds)
        # stopping criterion
        if play >= total_plays-1:
            break
            
    
    plt.plot(play_num,funds)
    final_funds.append(funds[-1])
    return final_funds
            
```
